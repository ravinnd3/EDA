## Check for distribution of columns

```python
import math
fitbit_df_cols = fitbit_df.columns.tolist()
fitbit_df_cols.remove('ActivityDate')

num_cols = 2  # Columns per row
num_rows = math.ceil(len(fitbit_df_cols) / num_cols)  # Dynamically calculate rows


plt.figure(figsize=(15, 5 * num_rows))  # Adjust figure size
for idx, col in enumerate(fitbit_df_cols):
    plt.subplot(num_rows, num_cols, idx + 1)
    plt.title(f"Histogram for {col} | Skewness: {round(fitbit_df[col].skew(),2)} ")
    sns.histplot(fitbit_df[col], bins='auto', cbar=True, kde=True)

plt.tight_layout()
plt.show()
```
